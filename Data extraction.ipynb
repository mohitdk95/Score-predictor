```python
import numpy as np
import pandas as pd
from yaml import safe_load
import os
from tqdm import tqdm
```

```python
filenames = []
for files in os.listdir('IPL data'):
    filenames.append(os.path.join('IPL data',files))
```

```python
filenames[0:5]
```

```python
df = pd.DataFrame()
counter = 1
for file in tqdm(filenames):
    with open(file,'r') as f:
        df1 = pd.json_normalize(safe_load(f))
        df1['match_id'] = counter
        df = df.append(df1)
        counter = counter + 1
        
```

```python
count = 1
deliver_df = pd.DataFrame()
for index, row in df.iterrows():
    if count in [75,108,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count +=1
        continue
    batting_team = []
    overs = []
    runs = []
    teams = []
    city = []
    venue = []
    match_id = []
    player_dismissed = []
    batsman = []
    bowler = []
    balls = []
    count +=1
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            runs.append(ball[key]['runs']['total'])
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            balls.append(key)
            try:
                player_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_dismissed.append('0')
    loop_df = pd.DataFrame({'match_id':match_id , 'teams':teams , 'batting_team':batting_team ,
                           'batsman':batsman , 'bowler':bowler , 'runs':runs , 'player_dismissed':player_dismissed,
                           'city':city , 'venue':venue , 'balls':balls})
    deliver_df = deliver_df.append(loop_df)
```

```python
df = deliver_df.copy()
```

```python
df.head()
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
df['bowling_team'] = df.apply(bowl , axis=1)
```
