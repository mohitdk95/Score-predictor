```python
df['current_score'] = df.groupby('match_id').cumsum()['runs']
```

```python
df['over'] = df['balls'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['balls'].apply(lambda x:str(x).split(".")[1])    
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
```

```python
df['balls_left'] = 120 - df['balls_bowled']
```

```python
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
df.head()
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
```

```python
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
df['wicket_left'] = 10 - df['player_dismissed']
```

```python
df.head(120)
```

```python
df['crr'] = df['current_score']*6 / df['balls_bowled']
```

```python
group = df.groupby('match_id')
match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(group.get_group(id).rolling(window = 30).sum()['runs'].values.tolist())
df['last_five'] = last_five
```

```python
x = df2[['current_score','balls_left','wicket_left','crr','last_five','batting_teamA','bowling_teamA']]
y = df2['runs_x']
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state = 1)
```

```python
from sklearn.ensemble import RandomForestRegressor
lin = RandomForestRegressor(n_estimators=100,max_features=None)
lin.fit(x_train,y_train)
```

```python
y_pred = lin.predict(x_test)
score = lin.score(x_test,y_test)
print("R-squared value:" , score)
# print("Custom accuracy:" , custom_accuracy(y_test,y_pred,20))
```

```python
trf = ColumnTransformer([('trf',OneHotEncoder(sparse=False , drop='first'),
                        ['batting_team','bowling_team','city'])],
                        remainder = 'passthrough')
pipe = Pipeline(steps=[('step1',trf),
                       ('step2',StandardScaler()),
                       ('step3',XGBRegressor(n_estimators = 1000 , learning_rate = 0.2 , 
                                            max_depth=12 , random_state = 1))])
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
print(r2_score(y_test , y_pred))
print(mean_absolute_error(y_test , y_pred))
```
